```python
# BANK MANNAGEMENT SYSTEM
print("*****WELCOME TO THE ABC BANK*****\n")
def Bank(Account_Details):
    print(("Enter Your Choice:\n"),
          ("1) New_Account\n"),
          ("2) Deposit\n"),
          ("3) Withdrawl\n"),
          ("4) Balance Enquiry\n"),
          ("5) Exit\n"))
    choice=int(input("Enter Your Choice:"))
    if choice >= 1 and choice <= 5:
        if choice == 1:
            print("...Provide Account Details...")
            account_holder_name = input("Enter A/C holder name:")
            pin = input("Enter PIN number(4 digits):")
            amount = float(input("Enter Depost Amount(min:500/-):"))
            if amount >= 500:
                Account_Details[account_holder_name] = {'pin': pin, 'amount': amount}
                print("\n Account Created Successfully....!!! \n")
                for name, details in Account_Details.items():
                    print(f"account_holder_name: {name}")
                    for key, value in details.items():
                        print(f"  {key.replace('_', ' ').title()}: {value}")
            else:
                print("Enter Minimum 500 rupees")


        elif choice == 2:
            account_holder_name = input("Enter A/C holder name:")
            pin = input("Enter PIN number:")
            if account_holder_name in Account_Details and Account_Details[account_holder_name]['pin'] == pin:
                print(" Your Current Balance:",Account_Details[account_holder_name]['amount'])
                Deposit_amount = float(input("Enter Deposit Amount:"))
                if Deposit_amount >= 0:
                    Account_Details[account_holder_name]['amount'] += Deposit_amount
                    print("Deposit Success..!!! \n")
                    print(" Updated Balance: \n", Account_Details[account_holder_name]['amount'])
                else:
                    print("Invalid Deposit Amount")
            else:
                print("Invalid Account Details")

        elif choice == 3:
            account_holder_name = input("Enter A/C holder name:")
            pin = input("Enter PIN number:")
            if account_holder_name in Account_Details and Account_Details[account_holder_name]['pin'] == pin:
                print(" Your Current Balance:",Account_Details[account_holder_name]['amount'])

                withdraw_amount = float(input("Enter Withdrawl Amount(min:500/-):"))
                if withdraw_amount>=500 and withdraw_amount <= Account_Details[account_holder_name]['amount']:
                    Account_Details[account_holder_name]['amount']=Account_Details[account_holder_name]['amount']- withdraw_amount
                    print("Withdrawl Successful...")
                    print("Updated Balance: \n", Account_Details[account_holder_name]['amount'])
                else:
                    print("Insufficient Balance, Check your Bank Balance...!!!")
            else:
                print(("Invalid Account Details \n"),
                      ("!!! Please check your A/C Holder Name and Pin !!!"))

        elif choice == 4:
            account_holder_name = input("Enter A/C holder name:")
            pin = input("Enter PIN number:")
            if account_holder_name in Account_Details and Account_Details[account_holder_name]['pin'] == pin:
                print("Current Balance: ", Account_Details[account_holder_name]['amount'],"\n")
            else:
                print("Invalid Account Details\n")

        elif choice == 5:
            print("Exit Application \n")
            print("Thank you for using the Bank Management System")
            return False
    else:
        print("Invalid Number, please Enter valid number")
    return True


Account_Details = {}
while True:
    if not Bank(Account_Details):
        break

    print("\nIf you want to try again")
    print("Enter Y to try again")
    print("Enter N to Stop")

    choose = input("Enter Y/N:").upper()
    if choose == 'Y':
        continue
    elif choose == 'N':
      print("Exit Application \n")
      print("Thank you for using the Bank Management System")
      break
    else:
        print("Invalid input. Please enter Y or N.\n")
```
